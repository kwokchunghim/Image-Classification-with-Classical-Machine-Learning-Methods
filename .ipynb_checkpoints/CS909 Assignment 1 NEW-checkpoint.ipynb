### 1.3 Performance Metrics

There are several performance metrics available for this kind of binary classification problem, e.g. accuracy, F1-score, Area Under Curve (AUC) of precision-recall (PR) curve, and Area Under Curve (AUC) of Receiver Operator Characteristics (ROC) curve. They are all designed to evaluate model performance from different perspectives, but we believe that not all of them are appropiate for our task as some of them do have a balanced data assumption.

Accuracy measures the overall correctness of the model. Since we have about 80% of the data coming from the same (negative) class, the model can be easily trained to achieve a 80% accuracy by simply predicting all the data to belong to the negative class. Hence, with such a heavy data imbalance in our dataset, a high prediction accuracy does not necesscarily implies that the model is well-trained. So, we believe that accuracy is not a suitable metric for our project.

ROC curves plot True Positive Rate (TPR) against False Positive Rate (FPR). With imbalanced data, AUC-ROC is not an appropiate metric for performance evaluation either. The formulas of FPR = FP / (FP + TN) and TPR = Recall = TP / (TP + FN) suggest that when we have a large number of True Negatives given a large number of negative examples, our FPR will be biasedly low. Also, ROC pays equal attention to prediction of both positive and negative classes, which is not the case in our task, as we are aim to train a model that can identify all the images of the positive class instead.


Therefore, we believe that AUC-PR would be the best option to evaluate our model performance. We note that the formulas of precision and recall are given by, respectively, Precision = TP / (TP + FP) and Recall = TP / (TP + FN). So, a large AUC-PR would require both our TP to be large enough relative to both FP and FN, i.e. maximizing # of TP predictions while minimizing # of FP and FN predictions, which aligns with our training objectives. So, it would be the most suitable metric to refer to during our training.

An alternative would be using F1-score, which is a function and a one-number summary of precision and recall that also tackles the problem of imbalanced data. 



### 1.4 Expected Accuracy of a Random Classifier

Consider a random binary classifier and randomly generates -1 or 1 outputs with probability = 0.5 each.

We argue that the expected accuracy of a completely random binary classifier (i.e. making a blind guess) would be 50% = 0.5, regardless of number of instances of the two data classes. Suppose n% of the data are of the positive class and (100 - n)% of the data are of the negative class. The expected accuracy = 0.5 * n + 0.5 * (1 - n) = 0.5, which does not depend on n.

We now run a tiny computer experiment to futher validate our argument.

```python
#We write a function to evaluate accuracy 

def accuracy(y1, y2):
    correct = 0
    wrong = 0

    for i in range(y1.shape[0]):
        if y1[i] == y2[i]:
            correct += 1
        else:
            wrong += 1
    return correct / (correct + wrong)


#Create an array of random '-1' and '1' of the same size of y_train 
y_random = np.array([-1 if random.uniform(-1, 1) < 0 else 1 for i in range(y_train.shape[0])]).reshape(y_train.shape)    
y_random2 = np.array([-1 if random.uniform(-1, 1) < 0 else 1 for i in range(y_train.shape[0])]).reshape(y_train.shape)    
y_random3 = np.array([-1 if random.uniform(-1, 1) < 0 else 1 for i in range(y_train.shape[0])]).reshape(y_train.shape)    



#Evaluate the accuracy
print(f"The accuracy of a random binary classifier applied on our training data is given by {accuracy(y_train, y_random) * 100:.{4}}%.")
print(f"The accuracy of another random binary classifier applied on our training data is given by {accuracy(y_train, y_random2) * 100:.{4}}%.")
print(f"The accuracy of the last random binary classifier applied on our training data is given by {accuracy(y_train, y_random3) * 100:.{4}}%.")
```

As expected, all the resulting accuracies are very close to 50% which is consistent with our argument above.

### 1.5 AUC-ROC and AUC-PR of a Random Classifier

We also argue that for a random classifier, the expected AUC-ROC would be 0.5 and the expected AUC-PR would be the proportion of positive examples, which is ~0.2 in our case. We write down a simple mathematical proof of our claims. Note that the AUC-ROC is the expectation, of TPR = TP / (TP + FN) over different FPR = FP / (TN + FP) values, and the AUC-PR is, on the other hand, an expectation of Precision = TP / (TP + FP) over different Recall = TP / (TP + FN) values. 

Suppose the total number of data = n, and 20% of them are of positive class. Suppose our random classifier is uniform with p = 0.5 predicting a positive class and p = 0.5 predicting a negative class, regardless of their true class. Here we note that the prediction of a random classifier is independent of the real labels, i.e. equivalent to a blind guess.

Hence, by independence of predicted and real labels,

Expected AUC-ROC = P(TP | P) = P(Predicted P)P(True P) / P(True P) = P(Predicted P) = 0.5 

Expected AUC-PR = P(TP | Predicted P) = P(Predicted P)P(True P) / P(Predicted P) = P(True P) = 0.2



```python
from sklearn.metrics import PrecisionRecallDisplay, precision_recall_curve, roc_curve, RocCurveDisplay, auc, average_precision_score

precision, recall, _ = precision_recall_curve(y_train, y_random)
fpr, tpr, _ = roc_curve(y_train, y_random)

roc_auc = auc(fpr, tpr)

display = PrecisionRecallDisplay(precision=precision, recall=recall)
display2 = RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc, estimator_name='ROC curve of Random Classifier')


display.plot()
display2.plot()

print(f"The AUC-PU (a.k.a. average precision score) of the Random Classifier is given by {average_precision_score(y_train, y_random):.{4}}.")
print(f"The AUC-ROC of the Random Classifier is given by {auc(fpr, tpr):.{4}}.")
print("\n")
print("TOP: PR Curve of Random Classifier; BOTTOM: ROC Curve of Random Classifier")
```

As we see, the results of computer experiment once again agrees with our theoretical result.

## 2. KNN Classifier

### 2.1 Stratified K-fold Cross Validation

We now train our initial model on a KNN Classfier (with k = 3) on the train dataset with stratified K-fold cross-validation (with K = 5) to evaluate our results. The reason why we need stratification is because of the class imbalance problem we mentioned above, so we wish to keep the relative percentage of each class in each of the folds. We keep track of the accuracy, balanced accuracy, AUC-ROC and AUC-PR as our performance metrics in our model evaluation and print the precision and the recall here as well for the ease of making analysis. As mentioned above, we should use AUC-PR as our main model evaluation metric.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import StratifiedKFold, cross_validate
from sklearn.metrics import ConfusionMatrixDisplay

#This will be reused in the later sections
skfold = StratifiedKFold(n_splits=5, random_state=42, shuffle=True)

clf = KNeighborsClassifier(n_neighbors=3)

#Note that 'average_precision' is essentially the AUC-PR estimated with numerical integration
metrics = ['accuracy', 'balanced_accuracy', 'average_precision', 'roc_auc', 'precision', 'recall']


def evaluate_model(X_train, y_train, metrics, skfold, pipeline):
    
    CV = cross_validate(pipeline, X_train, y_train, scoring=metrics, cv=skfold)

    for metric in metrics:
        print(f"Mean of {metric}: {np.mean(CV['test_' + metric]):.{4}}\nSD of {metric}: {np.std(CV['test_' + metric]):.{4}}")

print("Cross Validation results of 3-NN Classifier:")        
evaluate_model(X_train, y_train, metrics, skfold, clf)
```

Here, we present our outputs in a table below

| -----------          | :---------: | :---------:        |

| Balanaced Accuracy   | 0.7950      |  0.013440          |
| AUC-PR               | 0.6441      |  0.042990          |
| AUC-ROC              | 0.8830      |  0.014560         |


### 2.2 Plots and further investigation 

We observed that despite that we obtained a fairly high accuracy, the balanced accuracy, which is an average of sensitivity and specificity, is significantly lower. Also, our main performance metric, AUC-PR, is rather low compared to the other metrics. We suspect that it is due to the class imbalance of our data. To make further investigation, we present the confusion matrix, ROC curve and the PR curve of one of the folds in our cross-validation.

```python
#Confusion matrix for investigation

def plot_confusion(X_train, y_train):
    for _, (train, test) in enumerate(skfold.split(X_train, y_train)):
        clf.fit(X_train[train], y_train[train])
        ConfusionMatrixDisplay.from_estimator(clf, X_train[test], y_train[test])
        break

plot_confusion(X_train, y_train)      
```

### Observations

As we see, despite the model accurately predicted most of the true negative examples (which is not of our interest), both the precision and the recall is relatively low. A reasonable explanation of this phenomenon is that the large number of negative training examples 'encourages' our model to make more negative predictions. Also, it is also possible that we actually don't have enough positive samples to train our model.

We then plot the ROC curves and PR curves of the 3-NN classifier we fitted for one of the 5 folds.

```python
#ROC curve and PR curve

def plotROCPR(X_train, y_train):
    for _, (train, test) in enumerate(skfold.split(X_train, y_train)):
        proba = clf.fit(X_train[train], y_train[train]).predict_proba(X_train[test])
        RocCurveDisplay.from_predictions(y_train[test], proba[:, 1])
        PrecisionRecallDisplay.from_predictions(y_train[test], proba[:, 1])
    #     My original attempt without using predict_proba, gives the same result
    #     clf.fit(X_train[train], y_train[train])
    #     RocCurveDisplay.from_estimator(clf, X_train[test], y_train[test])
    #     PrecisionRecallDisplay.from_estimator(clf, X_train[test], y_train[test])
        break
        
plotROCPR(X_train, y_train)
print("TOP: PR Curve of 3-NN Classifier; BOTTOM: ROC Curve of 3-NN Classifier")
```

The result might not be optimal, but it is much better than the random classifier so this model does have certain predictive power.

### Conclusion: high accuracy does not imply good model performance

As expected, the ROC curve gives a much higher AUC than the PR curve does. This further validates our belief that the model attained a high accuracy mainly by giving True Negative Predictions, which is undesired. Apparently, we need to improve our model as its performance on predicting the positive examples, based on our metrics like precision and recall, is still mediocre. 

### 2.3 Impact of Preprocessing

As a first step to improve our model, we first make an attempt on data preprocessing and investigate how it impacts our validation results. Feature scaling is a common preprocessing as it enforces equal distribution in each feature. Standardization transforms each feature to Standard Normal Distribution and Min-Max Scaling scales each feature values to the range of [0,1]. We observe if these preprocessing improves our CV results.

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, Normalizer
from sklearn.pipeline import Pipeline

clf = KNeighborsClassifier(n_neighbors=3)
standard_pipe = Pipeline([('standardization', StandardScaler()), ('3-NN', clf)])
minmax_pipe = Pipeline([('min_max_normalization', MinMaxScaler()), ('3-NN', clf)])

print("KNN with Standardization:")
evaluate_model(X_train, y_train, metrics, skfold, standard_pipe)
print("\nKNN with Min-max Scaling:")
evaluate_model(X_train, y_train, metrics, skfold, minmax_pipe)
```

### Observation

We saw that both Min-Max Scaling and Standardization barely improves our results, especially on AUC-PR.
It is widely believed that feature-sclaing methods like standardization and min-max scaling does improve model performance of KNN, but its benefit to our model is very limited. One of the possible explanations is that our k=3 is too small for the feature scaling to work, as our model suffers from overfitting. It could be shown that if we increase k from 3 to 10, then standardization does slightly improve our result.

## 3. Cross Validation and Model Selection

In this section, we train our data using 4 different model architectures, namely, 1) Perceptron, 2) Gaussian Naive Bayes Classifier, 3) Linear SVM, and 4) Kernel SVM. Similar to what we did above, we use a 5-fold cross validation on our training data and evaluate the result using the same performance metrics. For each of the classifiers, we first apply grid search to do an exhausive search over the hyperparameter space to find the optimal hyperparameters for each classifier based on AUC-PR, i.e.'average_precision', with both raw and preprocessed (scaled) data. We then pick the best model from each classifier and make comparisons between them and present our results in a table.

```python
from sklearn.linear_model import Perceptron
from sklearn.naive_bayes import GaussianNB 
from sklearn.svm import LinearSVC, SVC
from sklearn.model_selection import GridSearchCV
```

```python
#2 different methods for preprocessing
scaler = StandardScaler()
scaler2 = MinMaxScaler()

X_train_standard = scaler.fit_transform(X_train)
X_test_standard = scaler.transform(X_test)

X_train_minmax = scaler2.fit_transform(X_train)
X_test_minmax = scaler2.transform(X_test)
```

```python
#function to cross validate with grid search with different preprocessed data
def model_cv(parameters, clf, metrics, X_train, y_train):

    clf_cv = GridSearchCV(clf, parameters, scoring='average_precision', cv=skfold, verbose=True)
    clf_cv.fit(X_train, y_train)

    clf_best = clf_cv.best_estimator_
    print(f"The best model given by grid search is {clf_best}")

    print(f"\nModel Performance of {clf_best} under CV:")
    evaluate_model(X_train, y_train, metrics, skfold, clf_best)
```

### 3.1 Perceptron

For our perceptron models, we apply grid search to search over different regularization constants, alpha, and different penalty norm. 

```python
#Perceptron

para_grid_perceptron = {'alpha':[0.01, 0.001, 0.0001, 0.00001, 0.000001], 'penalty':['l1', 'l2']}

perceptron = Perceptron(shuffle=True, random_state=42)

metrics = ['accuracy', 'balanced_accuracy', 'average_precision', 'roc_auc', 'precision', 'recall']

print("Without preprocessing:")
model_cv(para_grid_perceptron, perceptron, metrics, X_train, y_train)
print("\nStandardization:")
model_cv(para_grid_perceptron, perceptron, metrics, X_train_standard, y_train)
print("\nMin-Max Scaling:")
model_cv(para_grid_perceptron, perceptron, metrics, X_train_minmax, y_train)
```

We found that the best Perceptron model is given by setting alpha=1e-05 and penalty='l2' with Min-Max-Scaling preprocessing. As expected, the AUC-PR and overall model performance is not any better than the 3-NN classifier we fitted above, as our image data is unlikely to be linearly separable, which is an essential assumption of Perceptron.

### 3.2 Gaussian Naive Bayes Classifier 

The GNB classifier is a generative classification model based on probability, hence it is invariant to feature scaling like standardization. So, we apply GNB classifier to raw data and observe the result.

```python
#GNB Classifier

clf2 = GaussianNB()

metrics = ['accuracy', 'balanced_accuracy', 'average_precision', 'roc_auc', 'precision', 'recall']

print("GNB Classifier CV:")
evaluate_model(X_train, y_train, metrics, skfold, clf2)
```

### 3.3 Linear SVM

For our Linear SVM models, we again apply grid search to search for the optimal C, the regularization constant, and  different loss functions. We note that a higher C gives a more narrow boundary margin, thus gives reduces training error but also increases the risk of overfitting.  

```python
#Filter the long convergence warnings from Linear SVM
import warnings
warnings.simplefilter("ignore")
```

```python
para_grid_linearsvm = {'C':[0.125, 0.25, 0.5, 1, 2, 8], 'loss':['hinge', 'squared_hinge']}

linear_svm = LinearSVC(random_state=42, max_iter=3000)

metrics = ['accuracy', 'balanced_accuracy', 'average_precision', 'roc_auc', 'precision', 'recall']

print("Without preprocessing:")
model_cv(para_grid_linearsvm, linear_svm, metrics, X_train, y_train)
print("\nStandardization:")
model_cv(para_grid_linearsvm, linear_svm, metrics, X_train_standard, y_train)
print("\nMin-Max Scaling:")
model_cv(para_grid_linearsvm, linear_svm, metrics, X_train_minmax, y_train)
```

### 3.4 Kernel SVM

Finally for Kernel SVM, we apply grid search to search over different regularization parameters C, the kernel coefficients gamma, and the kernel type. Note that for RBF-SVM, a higher gamma gives a 'tighter' fit at a cost of increaseing the risk of overfitting.

```python
para_grid_kernelsvm = {'C': [0.125, 0.25, 0.5, 1, 2, 4, 8], 'gamma': ['auto', 'scale'], 'kernel': ['rbf', 'poly']}

kernel_svm = SVC(random_state=42)

metrics = ['accuracy', 'balanced_accuracy', 'average_precision', 'roc_auc', 'precision', 'recall']

print("Without preprocessing:")
model_cv(para_grid_kernelsvm, kernel_svm, metrics, X_train, y_train)
print("\nStandardization:")
model_cv(para_grid_kernelsvm, kernel_svm, metrics, X_train_standard, y_train)
print("\nMin-Max Scaling:")
model_cv(para_grid_kernelsvm, kernel_svm, metrics, X_train_minmax, y_train)
```

In terms of AUC-PR, the model with 'rbf' kernel, C = 4, and gamma = 'scale' performs the best in our cross-validation. It achieves 0.75 AUC-PR even before any preprocessing and outperforms any other classifiers in previous subsections and the standardization slightly boosts our AUC-PR to 0.766, which is by far the best-performing model.

### PR Curves 

We do the same for PR curves. Note that AUC-PR is the metric of our interests so more attention should be paid to this plot than the previous one.

```python
def plot_all_PR(X_train, y_train):
    fig = plt.figure()
    axes = fig.add_axes([1, 1, 1, 1])
    labels = {'1':'Perceptrons', '2':'GNB', '3':'Linear SVM', '4':'Kernel SVM'}
    for _, (train, test) in enumerate(skfold.split(X_train, y_train)):
        i = 1  #labels for legend
        for clf in all_clf:
            clf.fit(X_train[train], y_train[train])
            PrecisionRecallDisplay.from_estimator(clf, X_train[test], y_train[test], name=labels[str(i)], ax=axes)
            i += 1
        break
    plt.title("PR Curves of all classifiers for easy comparisons")

print('Top: Standardization; Bottom: Min-Max Scaling')
plot_all_PR(X_train_standard, y_train)
plot_all_PR(X_train_minmax, y_train)
```

### Observations

We saw that the Kernel SVM outperforms all the other classifiers with up to 0.92 AUC-ROC and 0.75 AUC-PR, in both standardization and min-max scaling settings, while the Gaussian NB classifier performed exceptionally poor, which is likely because of the wrong model assumption we mentioned. We believe that the image data are too complex for linear classification methods like Perceptrons, Linear SVM, as they are not likely be anywhere linearly separable. Kernel SVM, with the 'kernel trick', tackled the problem by mapping the data points to a high dimensional functional space to make classification.

We conclude that, out of the optimized classifiers we tested, Kernel SVM (with RBF kernel) performs the best and we will use it as a default classifier for our tasks in later sections.

## 4. Principal Component Analysis

Principal Component Analysis (PCA) is a common approach to reduce the dimension of our data based on Singular Value Decomposition (SVD). In particular, it exploits the k orthogonal directions that explain the most variance of the data as a low-dimensional representation. It is also useful for visualizing the distributions of high dimensional data. 

We first apply PCA to our training data and visualize the training data with a scatter plot to observe any significant difference between classes. Then we also apply PCA on both training and test data to observe the difference in distributions between training and test data. 

After that, we present the scree graph, which depicts the relationship the principal components and their corresponding proportion of variance explained. We make use of the proportion of variance explained to find the number of dimensions required to explain 95% of our data variance. Finally, we apply Kernel SVM on the transformed data and perform cross-validation on certain performance metrics to assess our results.

### Distribution of Training Data between Classes

We plot a figure to display the distribution of the training data points after applying PCA. Each data point of the postive class is represented by a red dot and point of the negative class is represented by a green square. We set alpha = 0.4 for high transperency to observe the overlapping data.

```python
from sklearn.decomposition import PCA

fig = plt.figure()
axes = fig.add_axes([1, 1, 1, 1])

pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train)

positive_idx = []
negative_idx = []

for i in range(X_train.shape[0]):
    if y_train[i] == 1:
        positive_idx.append(i)
    elif y_train[i] == -1:
        negative_idx.append(i)

axes.scatter(X_train_pca[positive_idx,0], X_train_pca[positive_idx,1], c='g', marker='s', alpha=0.4, label="Positive Class")
axes.scatter(X_train_pca[negative_idx,0], X_train_pca[negative_idx,1], c='r', marker='o', alpha=0.4, label="Negative Class")

plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')

plt.legend()
plt.show()

```

As depicted in the figure, there are huge overlaps between the data of positive class and the negative class. The negative class does seem to have smaller values in the first PC while the postive class tends to have a larger value in PC1, but there are plenty of exceptions behaving as noises that make our data nowhere near linearly separable (even after dimension reduction). This explains the poor performances of the linear classifiers in the previous section.

### Training Data vs Test Data

We do the same to compare the distribution between training data and test data after applying PCA to both. We observed during our data investigation that the training data contains much more noise that the test data so we have to see how the two distributions differ (after PCA).

```python
fig = plt.figure()
axes = fig.add_axes([1, 1, 1, 1])

pca2 = PCA(n_components=2)
X_test_pca = pca.fit_transform(X_test)

axes.scatter(X_train_pca[:,0], X_train_pca[:,1], c='b', alpha=0.4, marker='s', label='Training Data')
axes.scatter(X_test_pca[:,0], X_test_pca[:,1], c='y', alpha=0.4, marker='o', label='Test Data')

plt.legend()
plt.show()
```

We found that the training and the test data roughly follow the same distribution in the first two Principal Components and thus are comparable. We argue that there are two oblique gaps that seem to separate test data into three different groups, but they are not obvious and we believe that they are not too significant to our generalization. Our PCA visualization is consistent to our expectation based on data investigation. 

### Scree Plot and Proportion of Variance

The PCA scree graph plots the proportion of variance explained by each PC against the PCs (in a descending order based on variance explained). We observe that except for the first two PCs which totally explain more than 30% of the data variance, the rest of the PCs explain a relatively small proportion of variance each. This indicates that  we might still need a relatively large number of dimensions (PCs) to explain the amount of information contained in the data.

```python
fig = plt.figure()
axes = fig.add_axes([1, 1, 1, 1])

pca3 = PCA()
X_train_pca_all = pca3.fit_transform(X_train)
pc = np.arange(1, 31)

axes.plot(pc, pca3.explained_variance_ratio_[:30])


plt.title('Scree Graph of PCA')
plt.xlabel('PC number ordered by proportional of variance explained (descending)')
plt.ylabel('Proportion of variance explained')
plt.show()
```

We then find the number of dimensions needed to explain 95% of our training data variance.

```python
for i in range(X_train.shape[1]):
    if np.sum(pca3.explained_variance_ratio_[:i+1]) >= 0.95:
        print(f"The number of dimensions needed to explain 95% of the variance is given by {i + 1}.")
        break
```

We showed that we need only 314 dimensions (out of 768) to achieve explain 95% of the variance of our training data. This implies that more than half of the dimensions essentially contain little to none information. 

### Kernel SVM on data reduced by PCA

We now apply Kernel SVM on the data reduced by PCA. We choose n_components = 314 as it is the smallest number of dimensions needed to explain 95% of data variance. Similar to previous sections, we perform 5-fold stratified cross-validation and grid search to optimize the hyperparameters, and observe the impact of different data prepocessing techniques.

```python
pca4 = PCA(n_components=314)
X_train_pca_95 = pca4.fit_transform(X_train)
X_train_pca_standard95 = pca4.fit_transform(X_train_standard)
X_train_pca_minmax95 = pca4.fit_transform(X_train_minmax)

para_grid_pca = {'C': [0.25, 0.5, 1, 2, 4], 'gamma': ['scale', 'auto'], 'kernel': ['rbf', 'poly','sigmoid']}

clf = SVC(random_state=42)

metrics = ['accuracy', 'balanced_accuracy', 'average_precision', 'roc_auc', 'precision', 'recall']

print('Kernel SVM on PCA data')
model_cv(para_grid_pca, clf, metrics, X_train_pca_95, y_train)
print('\nKernel SVM on PCA data with standardization')
model_cv(para_grid_pca, clf, metrics, X_train_pca_standard95, y_train)
print('\nKernel SVM on PCA data with min-max scaling')
model_cv(para_grid_pca, clf, metrics, X_train_pca_minmax95, y_train)
```

We observed that, in terms of AUC-PR, the model performs almost on par with the Kernel PCA model before applying PCA. In particular, the Kernel SVM after PCA with standardization gives almost 0.76 AUC-PR, which is slightly lower than the original Kernel PCA. With the dimension reduction of PCA, we gained an advantage of a sigificant increase in training speed (i.e. computation efficiency), as the number of dimensions of data being processed during training is reduced by almost 60%.

Dimensional reduction in general does increase model interpretability and training efficiency at a cost of loss of data spatial information, but here we showed that the 'information loss' are not significant to making predictions at all. We believe that the PCA, in fact, , denoises our training data by discarding the redundant dimensions. Hence we conclude that our Kernel SVM after PCA gives us a better model in terms of computational efficiency without compromising on model performance.



## 5. Optimal Pipeline for future Predictions

Based on our analysis from Section 1 to 4, the Kernel PCA with RBF kernel with C = 4, gamma='scale', and standardized data gives the best model based on AUC-PR (AP = 0.766). We also saw that applying PCA to our data increases our model computational efficiency without compromising our model performance, so it seems to be an ideal option too. In this section, we attempt to further optimize this model and develop a pipeline to make predictions on unseen (test) data. 

In particular, the PCA dimension that explains 95% of the variance is the dimension that applies before the standardization. The number of dimension needed to explain the same amount of variance on standardized data could be different.

```python
pca3 = PCA()
X_train_pca_all = pca3.fit_transform(X_train_standard)

for i in range(X_train.shape[1]):
    if np.sum(pca3.explained_variance_ratio_[:i+1]) >= 0.95:
        print(f"The number of dimensions needed to explain 95% of the variance is given by {i + 1}.")
        break
```

```python
pca5 = PCA(n_components=362)
X_train_pca_standard_95 = pca5.fit_transform(X_train_standard)

para_grid_pca = {'C': [0.25, 0.5, 1, 2, 4], 'gamma': ['scale', 'auto'], 'kernel': ['rbf', 'poly','sigmoid']}

clf = SVC(random_state=42)

metrics = ['accuracy', 'balanced_accuracy', 'average_precision', 'roc_auc', 'precision', 'recall']

print('Kernel SVM on PCA standardized data with updated dimensions:')
model_cv(para_grid_pca, clf, metrics, X_train_pca_standard_95, y_train)
```

The model performance slightly improve with PCA dimension = 362 in terms of AUC-ROC. We adopt the following as our optimal pipeline to make model predictions.

```python
opt_pipe = Pipeline([('scaler', StandardScaler()), 
                 ('pca', PCA(n_components=362)), 
                 ('kernel_svm', SVC(C=2, random_state=42))])

opt_pipe.fit(X_train_standard, y_train)

#This will be our model output for test data
opt_pipe.predict(X_test_standard)
```

Before creating a textfile with all the predictions, we final check our model predictions on the training data to ensure that the pipeline works as we expected.

```python
#For pipeline reliability purpose only, NOT FOR PREDICTION/EVALUATION!!!!!

x = opt_pipe.predict(X_train_standard)

TP = 0
FP = 0
TN = 0
FN = 0

for i in range(len(y_train)):
    if x[i] == y_train[i] == 1:
        TP += 1
    elif x[i] == y_train[i] == -1:
        TN += 1
    elif x[i] == 1 and y_train[i] == -1:
        FP += 1
    elif x[i] == -1 and y_train[i] == 1:
        FN += 1

#Ensure the results makes sense - total number training data = 3000, with 611 positive and 2389 negatives
print(f"Training TP: {TP}")
print(f"Training FP: {FP}")
print(f"Training TN: {TN}")
print(f"Training FN: {FN}")
```
